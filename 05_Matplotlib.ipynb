# Matplotlib

`matplotlib.pyplot` is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

The illustration below contains the name of several matplotlib elements composing a figure:

<img src="00 - figure anatomy.png">

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 100)
y1 = np.random.normal(0, 10, 100)
```

`matplotlib.pyplot.plot`: Plot y versus x as lines and/or markers.

```python
plt.plot(x, y1, label='normal')
```

Plotting a second set of points:

```python
y2 = np.random.exponential(10, 100)
plt.plot(x, y1, label='normal')
plt.plot(x, y2, label='exponential')
```

Plotting a third set of points:

```python
y1 = np.random.normal(0, 10, 100)
y2 = np.random.exponential(10, 100)
y3 = np.random.binomial(50, 0.2, 100)

plt.plot(x, y1, label='normal')
plt.plot(x, y2, label='exponential')
plt.plot(x, y3, label='binomial')
```

`matplotlib.pyplot.xlabel`: Set the label for the x-axis.

```python
plt.plot(x, y1, label='normal')
plt.xlabel('x points')
```

`matplotlib.pyplot.ylabel`: Set the label for the y-axis.

```python
plt.plot(x, y1, label='normal')
plt.ylabel('normal points')
```

`matplotlib.pyplot.title`: Set a title for the axes.

```python
plt.plot(x, y1, label='normal')
plt.title('random normal points')
```

`matplotlib.pyplot.legend`: Place a legend on the axes.

```python
plt.plot(x, y1, label='normal')
plt.plot(x, y2, label='exponential')
plt.plot(x, y3, label='binomial')

plt.xlabel('x points')
plt.ylabel('normal points')
plt.title('random normal points')
plt.legend()
```

`matplotlib.pyplot.savefig(<filename>)`: Save the current figure to a file.

```python
plt.plot(x, y1, label='normal')
plt.plot(x, y2, label='exponential')
plt.plot(x, y3, label='binomial')

plt.xlabel('x points')
plt.ylabel('normal points')
plt.title('random normal points')
plt.legend()

plt.savefig('sample plot.png')
```

Matplotlib allows having multiple axes (plots) on a figure. All plotting commands apply to the current axes.

`matplotlib.pyplot.subplots(<nrows=1>, <ncols=1>)`: Create a figure and a set of subplots.

`matplotlib.pyplot.subplot(<nrows>, <ncols>, <index)>`: Add/select a subplot to the current figure.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.rand(1000)
y.sort()
x = np.arange(len(y))
```

```python
fig, ax = plt.subplots(2, 2)

# linear
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)


# logit
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.yscale('logit')
plt.minorticks_off()
plt.title('logit')
plt.grid(True)
```

`matplotlib.pyplot.subplots_adjust(left=<>, bottom=<>, right=<>, top=<>, wspace=<>, hspace=<>)`: Tune the subplot layout.

`left`: the left side of the subplots of the figure
`right`: the right side of the subplots of the figure
`bottom`: the bottom of the subplots of the figure
`top`: the top of the subplots of the figure
`wspace`: the amount of width reserved for space between subplots, expressed as a fraction of the average axis width
`hspace`: the amount of height reserved for space between subplots, expressed as a fraction of the average axis height

```python
fig, ax = plt.subplots(2, 2)

# linear
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)


# logit
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.yscale('logit')
plt.minorticks_off()
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top=5.92, bottom=5.08, left=0.10, right=0.95, hspace=1.35,
                    wspace=1.35)
```

# Bar Chart

`matplotlib.pyplot.bar(x, height)`: Make a bar plot.
`x`: sequence of scalars
`height`: scalar or sequence of scalars

`matplotlib.pyplot.xticks(ticks=<>, labels=<>)`: Get or set the current tick locations and labels of the x-axis.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
x_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
```

```python
plt.bar(x_pos, performance, color='green')

plt.xticks(x_pos, people)

plt.ylabel('Performance')
plt.title('How fast do you want to go today?')

plt.show()
```

```python
men_means = (20, 35, 30, 35, 27)
women_means = (25, 32, 34, 20, 25)

indices = np.arange(len(men_means))  # the x locations for the groups
width = 0.35  # the width of the bars
```

Multiple data sets can be added to a bar chart. The key is to adjust their location on the x-axis properly, so they are displayed next to each other.

```python
plt.bar(indices - width/2, men_means, width, color='SkyBlue', label='Men')
plt.bar(indices + width/2, women_means, width, color='IndianRed', label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(indices, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35          # the width of the bars: can also be len(x) sequence
```

To create stacked charts (bar charts stacked on top of each other), the bottom argument of the bar function can be adjusted.

```python
p1 = plt.bar(ind, menMeans, width)
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
```

# Pie Chart

Pie chart, where the slices will be ordered and plotted counter-clockwise.

`matplotlib.pyplot.pie(x, explode=<>, labels=<>, colors=<>, autopct=<>):` Plot a pie chart.
`labels`: a sequence of strings providing the labels for each wedge.
`autopct`: used to label the wedges with their numeric value.
`explode`: specifies the fraction of the radius with which to offset each wedge.

`matplotlib.pyplot.axis`: Convenience method to get or set some axis properties. The 'equal' value sets equal scaling (i.e., make circles circular) by changing axis limits.

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
```

```python
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
```

```python
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```

```python
size = 0.3
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])
```

Matplotlib provides several color maps, which can be used to change the default plot colors. They can be found in the link below:
https://matplotlib.org/examples/color/colormaps_reference.html

```python
cmap = plt.get_cmap('tab20c')
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))
```

We can make donut charts (a special type of pie chart), by overlaying pie charts. We can do this using additional parameters of the pie function:
`radius`: The radius of the pie, if radius is None it will be set to 1.
`pctdistance`: The ratio between the center of each pie slice and the start of the text generated by autopct.
`labeldistance`: The radial distance at which the pie labels are drawn.
`wedgeprops`: Dict of arguments passed to the wedge objects making the pie. For example, width determines the width of the wedge (from outer radius to inner radius) and edgecolor sets the patch edge color.

```python
plt.pie(vals.sum(axis=1), radius=1, labels=['US', 'Canada', 'Mexico'], autopct='%1.0f%%', pctdistance=0.85,
       wedgeprops=dict(width=size, edgecolor='w'), colors=outer_colors)

plt.pie(vals.flatten(), radius=1-size, autopct='%1.0f%%', pctdistance=0.8,
       labeldistance=0.64, wedgeprops=dict(width=size, edgecolor='w'), colors=inner_colors)

plt.title('Pie plot with `plt.pie`')
plt.axis('equal')
```

## Scatter Plot

`matplotlib.pyplot.scatter(<x>, <y>)`: A scatter plot of `y` vs `x` with varying marker size and/or color.

```python
x = np.random.randn(1000)
y = np.random.randn(1000) - 1
z = np.random.randn(1000) + 1
```

```python
plt.scatter(x, y, marker='.', c='r')
plt.scatter(x, z, marker='.')
```

In some scenarios, if the axis limits are not appropriate, we can set them manually.

`matplotlib.pyplot.xlim`: Get or set the x-limits of the current axes.

`matplotlib.pyplot.ylim`: Get or set the y-limits of the current axes.

```python
plt.scatter(x, y, marker='.', c='r')
plt.scatter(x, z, marker='.')

binwidth = 0.25
xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
lim = (int(xymax/binwidth) + 1) * binwidth

plt.xlim((-lim, lim))
plt.ylim((-lim, lim))
```

```python
N = 200

x = np.random.rand(N)
x.sort()

y = np.random.rand(N)
y.sort()
y = y + np.random.randn(N) / 4

colors = np.random.rand(N)
colors.sort()

area = (30 * np.random.rand(N))**2  # 0 to 15 point radii
```

Size and color of the points on the scatter plot can be modified using the following arguments of the scatter function:
`s`: The marker size in points**2.
`c`: The marker color.
`alpha`: The point transparency, between 0 (transparent) and 1 (opaque).

```python
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```

```python
# sample data
N = 500
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(N)
```

```python
# we can specify number of the bins on the histogram
num_bins = 50
```

`matplotlib.pyplot.hist`: Plot a histogram.

```python
import matplotlib.pyplot as plt
plt.hist(x, num_bins, density=1)
```

We can add the "best fit" line as follows:

```python
n, bins, patches = plt.hist(x, num_bins, density=1)
# hist function returns some values, including bins, which contains the ranges for each bin
# we use this to generate our best fit line segments

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
plt.plot(bins, y, '--')
plt.xlabel('Smarts')
plt.ylabel('Probability density')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
```
